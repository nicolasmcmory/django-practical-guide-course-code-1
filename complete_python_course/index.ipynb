```python
nums = [1,2,3]
new_list = list(map(lambda item: item * 2, nums))
print(new_list)
```

```python
def new_list_func(func) -> list:
    def make_list(*args, **kwargs):
        return list(func(*args, **kwargs))

    return make_list


@new_list_func
def get_pairs(lst: list[int]) -> list[int]:
    return filter(lambda item: item % 2 == 0, lst)


get_pairs([1, 2, 3, 4, 5, 6])
```

```python
lst_1 = [1, 2, 3]
lst_2 = ["a", "b", "c"]

@new_list_func
def zip_it(list1,list2):
    return zip(list1,list2)

@new_list_func
def zip_it2(*args):
    return zip(*args)

print(zip_it(lst_1, lst_2))
print(zip_it2(lst_1, lst_2))
```

```python
def add(a, b):
    return a + b

params = (2, 3)
print(params)
print(*params)
print(add(*params))  # Output: 5

kwparams = {'a': 2, 'b': 3}
print(add(**kwparams))  # Output: 5
```

```python
from functools import reduce

nums = [1,2,3]

average = reduce(lambda reducer, item: reducer + item,nums, 0)/len(nums)

print(average)
```

```python
tuples = [(1, 2), (-1, 1), (5, 2)]
new_tups = sorted(tuples, key=lambda k: k[1])

print(tuples)
print(new_tups)
```

```python
some_list = ['a', 'b', 'c', 'b', 'd', 'm', 'n', 'n']

duplicates = []
for value in some_list:
    if some_list.count(value) > 1:
        if value not in duplicates:
            duplicates.append(value)

print(duplicates)
```

```python
dup = list({item for item in some_list if some_list.count(item) > 1})

dup = filter(lambda item: some_list.count(item) > 1, some_list)

print(list(dup))
```
